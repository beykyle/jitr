```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import special as sc


def spherical_scattering_wave_function(r, theta, l, eta):
    return (
        np.absolute(
            eta
            * r
            * sc.spherical_jn(l, eta * r)
            * sc.sph_harm_y(l, 0, np.pi / 2 + theta, 0)
        )
        ** 2
    )


l = 1
eta = 3
r = np.linspace(0.01, 2 * np.pi, 1000)
theta = np.linspace(0.001, 2 * np.pi, 1000)

R, Theta = np.meshgrid(r, theta)
Z = spherical_scattering_wave_function(R, Theta, l, eta)
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

fig = plt.figure(figsize=(8, 8), dpi=1000)
fig.patch.set_alpha(0.0)
plt.pcolormesh(X, Y, Z, shading="auto", cmap="YlOrBr_r", vmax=0.4, vmin=0.00)

plt.text(
    3,
    -0.5,
    r"$\mathcal{R}$",
    fontsize=300,
    horizontalalignment="center",
    verticalalignment="center",
    color="w",
)
plt.text(
    -2.5,
    0.0,
    r"jit",
    fontsize=200,
    horizontalalignment="center",
    verticalalignment="center",
    color="w",
)
plt.yticks([])
plt.xticks([])
plt.axis("off")
plt.tight_layout()
plt.savefig("jitr_logo.png")
```
